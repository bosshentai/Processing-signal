<p> declare um eixo temporal discreto t de dimensao N = 1024 para t1 = 0,..., tn = 2 segundos, utilizando a função linspace. Qual o intervalo de tempo entre cada amostra em segundos ? Procure um outra forma de fazer o mesmo eixo temporal sem utilizar a função linspace. E se quiser fazer um eixo tempora de [-2,2] segundos ? </p>

```python
N = 1024 # amostras
t1 = 0 # tempo inicial 
tn = 2 # tempo final 
tempo_1 = np.linspace(t1,tn,N)
#tempo_1.dtype
#len(tempo_1)
intervalo_amostra_1 = tempo_1[2] - tempo_1[1]
intervalo_amostra_1
```

```python
# outra forma de fazer  o eixo temporal 
tempo_teste = np.arange(t1,tn,2/N)
len(tempo_teste)
```

```python
t1_2 = -2 
tn_2 = 2 
tempo_2 = np.linspace(t1_2,tn_2 , N)
```

<p> 2. construa agora uma funcao s(t) = sin(x) onde x = w0*t.Se quisermos um seno de frequencia f0 = 10 Hz, como fazer? Construa o vector s em funcao do vector t do numero anterior. </p> 

```python
def st(frequencia ,tempo):
    x = 2*(math.pi)*frequencia*tempo
    x = np.sin(x)
    return x
```

```python
f0 = 10
st_1 = st(f0,tempo_1)
st_2 = st(f0,tempo_2)
```

<p>3. Queremos aumentar a amplitude da função s(t) para um valor A = 4.3 </p>

```python
ampl = 4.3
st_1_amo = ampl * st_1
st_2_amo = ampl * st_2
```

<p> 4. vamos agora  construir uma nova funcao g(t)= exp(- alpha * t ) onde alpha = 2 . Determine o vector g .

```python
def gt(alpha,tempo):
    teste = (-alpha * tempo)
    x = np.exp(tempo)
    return x 
```

```python
alpha = 2 
gt_1 = gt(alpha,tempo_1)
gt_2 = gt(alpha,tempo_2)
```

<p> 5. utilizando a função plot represente os vectores s  e g , primeior em duas figuras separadas, depois no mesmo grafico e a função subplot. Coloque  legendas nos eixos  do tempo e amplitude e titulos no varios grafico </p>

```python
# subplot 
plt.figure(figsize=(20,12))
ax1 = plt.subplot(2,2,1)
#ax1.margins(x=0,y=0)
ax1.plot(tempo_1,st_1_amo)
ax1.set_title('s(t)')
ax1.set_ylabel('Amplitude')
ax1.set_xlabel('Tempo')
#ax1.set_ylim(min(st_1_amo),max(st_1_amo))

ax2 = plt.subplot(2,2,2)
ax2.plot(tempo_1,gt_1)
ax2.set_title('g(t)')
ax2.set_ylabel('Amplitude')
ax2.set_xlabel('Tempo')
#ax2.set_ylim(min(gt_1),max(gt_1))
#ax2.set_xlim(min(tempo_1),max(tempo_1))

ax3 = plt.subplot(2,2,3)
ax3.plot(tempo_2,st_2_amo)
ax3.set_title('s(t)')
ax3.set_ylabel('Amplitude')
ax3.set_xlabel('Tempo')
#ax3.set_ylim(min(st_2_amo),max(st_2_amo))
#ax3.set_xlim(min(tempo_2),max(tempo_2))

ax4 = plt.subplot(2,2,4)
ax4.plot(tempo_2,gt_2)
ax4.set_title('g(t)')
ax4.set_ylabel('Amplitude')
ax4.set_xlabel('Tempo')
plt.show()
```
